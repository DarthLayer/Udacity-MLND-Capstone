```python
df['close_1daylag'] = df.groupby('symbol')['close'].shift(1)
df['close_5daylag'] = df.groupby('symbol')['close'].shift(5)
df['close_10daylag'] = df.groupby('symbol')['close'].shift(10)
df['close_30daylag'] = df.groupby('symbol')['close'].shift(30)
df['close_60daylag'] = df.groupby('symbol')['close'].shift(60)
df['close_180daylag'] = df.groupby('symbol')['close'].shift(180)
```

```python
df['close_1dayreturn'] = (df['close']-df['close_1daylag'])/df['close_1daylag']
```

```python
df.head()
```

```python
df['30day_mvgavg'] = df.groupby('symbol').close.apply(lambda x: x.ewm(alpha=1/30,min_periods=30).mean())
```
